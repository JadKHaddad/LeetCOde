```python
from typing import List

def show(board):
    for i in range(0, len(board)):
        for j in range(0, len(board[i])):
            print(board[i][j] , " ", end="")
        print("\n")
```

```python
class Solution:
    def split(self, word):
        return [char for char in word]

    def exist(self, board: List[List[str]], word: str) -> bool:
        if len(word) < 1:
            return True
        n = len(board)
        m = len(board[0])
        if len(word) > n*m:
            return False
        def test_dude(dude, letters) -> bool:
            if len(letters) < 1:
                return True
            i = dude[0]
            j = dude[1]
            letters_ = letters.copy() 
            if i + 1 < n:
                if board[i+1][j] == letters_[0]:
                    cell = board[i+1][j]
                    board[i+1][j] = "0"
                    letters_.remove(cell)
                    if test_dude((i+1,j), letters_):
                        return True
                    letters_.insert(0,cell)
                    board[i+1][j] = cell
            if not i - 1 < 0:
                if board[i-1][j] == letters_[0]:
                    cell = board[i-1][j]
                    board[i-1][j] = "0"
                    letters_.remove(cell)
                    if test_dude((i-1,j), letters_):
                        return True
                    letters_.insert(0,cell)
                    board[i-1][j] = cell
            if j + 1 < m:
                if board[i][j+1] == letters_[0]:
                    cell = board[i][j+1]
                    board[i][j+1] = "0"
                    letters_.remove(cell)
                    if test_dude((i,j+1), letters_):
                        return True
                    letters_.insert(0,cell)
                    board[i][j+1] = cell
            if not j - 1 < 0:
                if board[i][j-1] == letters_[0]:
                    cell = board[i][j-1]
                    board[i][j-1] = "0"
                    letters_.remove(cell)
                    if test_dude((i,j-1), letters_):
                        return True
                    letters_.insert(0,cell)
                    board[i][j-1] = cell
            return False

        letters = self.split(word)
        for i in range(0, n):
            for j in range(0, m):
                letters_ = letters.copy()
                cell_ = board[i][j]
                if cell_ == letters_[0]:
                    board[i][j] = "0"
                    letters_.remove(cell_)
                    if len(letters_) < 1:
                        return True
                    if i + 1 < n:
                        if board[i+1][j] == letters_[0]:
                            cell = board[i+1][j]
                            board[i+1][j] = "0"
                            letters_.remove(cell)
                            if test_dude((i+1,j), letters_):
                                return True
                            letters_.insert(0,cell)
                            board[i+1][j] = cell
                    if not i - 1 < 0:
                        if board[i-1][j] == letters_[0]:
                            cell = board[i-1][j]
                            board[i-1][j] = "0"
                            letters_.remove(cell)
                            if test_dude((i-1,j), letters_):
                                return True
                            letters_.insert(0,cell)
                            board[i-1][j] = cell
                    if j + 1 < m:
                        if board[i][j+1] == letters_[0]:
                            cell = board[i][j+1]
                            board[i][j+1] = "0"
                            letters_.remove(cell)
                            if test_dude((i,j+1), letters_):
                                return True
                            letters_.insert(0,cell)
                            board[i][j+1] = cell
                    if not j - 1 < 0:
                        if board[i][j-1] == letters_[0]:
                            cell = board[i][j-1]
                            board[i][j-1] = "0"
                            letters_.remove(cell)
                            if test_dude((i,j-1), letters_):
                                return True
                            letters_.insert(0,cell)
                            board[i][j-1] = cell
                board[i][j] = cell_
        return False
```
